```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from numpy import reshape
import matplotlib.pyplot as plt

# Step 1: Load the dataset
url = "https://raw.githubusercontent.com/ereilly89/chicago-housing-rental-app/master/datasets/listingTrainingData_Header.csv"
training_data = pd.read_csv(url)

training_data
```

```python
# Step 2: Preprocess the data
X = training_data.drop("Private room", axis=1)
y = training_data["Private room"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)

```

```python
# Step 3: Build the model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
# Step 4: Make predictions
y_pred = model.predict(X_test)
```

```python
# Step 5: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
meansqrd = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("Accuracy:", accuracy)
print("Mean Squared Error", meansqrd)
print("R2 score", r2)
```

```python
# Plotting the results
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Accuracy plot
ax[0].bar(['Accuracy'], [accuracy])
ax[0].set_ylabel('Accuracy')
ax[0].set_ylim(0, 1)
ax[0].set_title('Model Accuracy')

# Mean Squared Error plot
ax[1].bar(['Mean Squared Error'], [meansqrd])
ax[1].set_ylabel('Mean Squared Error')
ax[1].set_title('Mean Squared Error')

# R2 Score plot
ax[2].bar(['R2 Score'], [r2])
ax[2].set_ylabel('R2 Score')
ax[2].set_ylim(-1, 1)
ax[2].set_title('R2 Score')

plt.tight_layout()
plt.show()
```
